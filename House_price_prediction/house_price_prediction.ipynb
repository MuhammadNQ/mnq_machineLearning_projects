## Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a Pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
# Print First 5 rows of our DataFrame
house_price_dataframe.head()
```

```python
house_price_dataframe['MedHouseVal'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and Columns in the data frame
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

## Understanding the correlation between various features in the dataset

1. Positive Correlation
2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to nderstand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

## Splitting the data and Target

```python
X = house_price_dataframe.drop(['MedHouseVal'], axis=1)
Y = house_price_dataframe['MedHouseVal']
```

```python
print(X)
print(Y)
```

## Splitting the data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

### XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

## Evaluation

### Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
R_squared_score = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
mean_absolute_score = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", R_squared_score)
print('Mean Absolute Error : ', mean_absolute_score)
```

## Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Median house value")
plt.ylabel("Predicted Median house value")
plt.title("Actual Median house value vs Preicted Median house value")
plt.show()
```

## Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
r_squared_score = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
mean_absolute_score = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", r_squared_score)
print('Mean Absolute Error : ', mean_absolute_score)
```
